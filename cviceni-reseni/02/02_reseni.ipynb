## Titanic

V souboru [titanic.csv](data/titanic.csv) najdeš informace o cestujících na zaoceánském parníku Titanic. Vytvoř kontingenční tabulku, která porovná závislost mezi pohlavím cestujícího (sloupec `Sex`) a tím, jesti přežil potopení Titanicu (sloupec `Survived`). Dále vytvoř další kontingenční tabulku, která porovná závislost mezi třídou (sloupec Pclass), ve které cestoval, a tím, jestli přežil. Zkus spočítat relativní počty u jednotlivých tabulek.

```python
# Načtení dat
import pandas

# Modul pandas lze načíst i pod zvoleným aliasem (výhodou je pak kratší zápis, např. pd.read_csv() místo pandas.read_csv())
# (pro pandas se obvykle ze zvyku používá pd)
# import pandas as pd

data = pandas.read_csv("data/titanic.csv")
data.head()
```

```python
# Bonus: Pro přehlednost přidáme před číslo třídy řetězec "Class"

# 1. způsob:
# df["Pclass"] = "Class " + df["Pclass"].astype(str)

# 2. způsob:
data["Pclass"].replace({1:"Class 1", 2:"Class 2", 3:"Class 3"}, inplace=True)

data.head()
```

* Vytvoř kontingenční tabulku, která porovná závislost mezi pohlavím cestujícího (sloupec `Sex`) a tím, jestli přežil potopení Titanicu (sloupec `Survived`).

```python
# Vytvoření kontingenční tabulky (absolutní počet přeživších - podle pohlaví)
titanic_pivot_table1 = pandas.pivot_table(data, values="Name", index="Sex", columns="Survived", aggfunc="count", margins=True)
titanic_pivot_table1
```

* Dále vytvoř další kontingenční tabulku, která porovná závislost mezi třídou (sloupec Pclass), ve které cestoval, a tím, jestli přežil.

```python
# Vytvoření kontingenční tabulky (absolutní počet přeživších - podle třídy)
titanic_pivot_table2 = pandas.pivot_table(data, values="Name", index="Pclass", columns="Survived", aggfunc="count", margins=True)
titanic_pivot_table2
```

* Zkus spočítat relativní počty u jednotlivých tabulek.

```python
# display(titanic_pivot_table1)
# ...Funkce display() umožní vždy zobrazit DataFrame - jinak se v Jupyter notebooku zobrazuje jen poslední DF v buňce

# Tabulka s relativními poměry přeživších ve skupinách (podle pohlaví)
# (Všechny sloupce vydělíme posledním sloupcem, čímž získáme procenta)
titanic_pivot_table1_relative = titanic_pivot_table1.div(titanic_pivot_table1.iloc[:,-1], axis=0)
titanic_pivot_table1_relative
```

```python
# display(titanic_pivot_table2)

# Tabulka s relativními poměry přeživších ve skupinách (podle třídy)
titanic_pivot_table2_relative = titanic_pivot_table2.div(titanic_pivot_table2.iloc[:,-1], axis=0)
# titanic_pivot_table2_relative = titanic_pivot_table2.div(titanic_pivot_table2.iloc[-1,:], axis=1)

titanic_pivot_table2_relative
```

Bonus: Alternativní způsoby struktury pivot table...

```python
# Víceúrovňová struktura pivot table
# Výstup: relativní počet přeživších vzhledem k celkovému počtu pasažérů
# (pro zajímavost, pivot_table lze volat přímo jako metodu DataFramu, tedy data.pivot_table(...) místo pandas.pivot_table(x=data, ...))
print("Alternative pivoting - survival rate (%) of each group:")

df_pivot_alt1 = data.pivot_table(index=["Pclass","Sex"], values="Survived", aggfunc="mean", margins=True) *100
df_pivot_alt1
```

```python
df_pivot_alt2 = data.pivot_table(index=["Sex","Pclass"], values="Survived", aggfunc="mean", margins=True) *100
df_pivot_alt2
```

```python
df_pivot_alt3 = data.pivot_table(index="Sex", columns="Pclass", values="Survived", aggfunc="mean", margins=True) *100
df_pivot_alt3
```

* Dále využij metodu `cut` na rozdělení cestujících do věkových skupin (zkus vytvořit např. 4 skupiny, můžeš definovat hranice skupin tak, aby vznikly skupiny děti, teenageři, dospělí a senioři). Urči počet přeživších pro jednotlivé věkové skupiny.

```python
# Do DataFramu "data" přidáme sloupec "age_group" s rozdělením do skupin podle věku
data["age_group"] = pandas.cut(x=data["Age"], bins=[0, 12, 20, 60, float("inf")], labels=["Children", "Teenagers", "Adults", "Elderly"])
data.head(20)
```

```python
# Absolutní počty (ne)přeživších podle skupin
titanic_pivot_table_groups = pandas.pivot_table(data, values="Name", index="age_group", columns="Survived", aggfunc="count", margins=True)
titanic_pivot_table_groups
```

```python
# Bonus: Relativní počet (procenta) přeživších vzhledem ke všem pasažérům
# Funkce "mean" vyprůměruje záznamy "0" (nepřežil) a "1" (přežil), čímž získáme procento přeživších
titanic_pivot_table_groups_relative_to_all = data.pivot_table(index="age_group", columns="Sex", values="Survived", aggfunc="mean", margins=True) *100
display(titanic_pivot_table_groups_relative_to_all)
```

```python
# Bonus 2: Relativní poměry přeživších ke skupinách
titanic_pivot_table_groups_relative = titanic_pivot_table_groups.div(titanic_pivot_table_groups.iloc[:,-1], axis=0) *100
display(titanic_pivot_table_groups_relative)
```
